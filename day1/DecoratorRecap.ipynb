```python
# instrumentation example
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result
    return wrapper

@my_decorator
def add(a, b):
    return a + b

print(add(3, 4))
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Time taken: {end - start:.4f}s")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(1)

slow_function()
```

```python
# check conditions, conditional execution like auth token, expiry check

def requires_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Only positive numbers allowed")
        return func(x)
    return wrapper

@requires_positive
def square(x):
    return x * x

print(square(5))
# square(-3) â†’ error
```
